```python
def estatisticas(media, quartil=False):
    # data= {
    #     "Soma": [media.sum()], 
    #     "Media": [media.mean()],
    #     "Min": [media.min()],
    #     "Max":[media.max()],
    #     "Media":[media.mean()],
    #     "Mediana":[media.median()],
    #     "Coef_Var":[media.std(ddof=0)/media.median()],
    #     "Vari_N":[media.var(ddof=0)],
    #     "Vari_N-1":[media.var()],
    #     "DesvPad_N":[media.std(ddof=0)],
    #     "DesvPad_N-1":[media.std()],
    #     "25%": [media.quantile(q=0.25)],
    #     "50%": [media.quantile(q=0.50)],
    #     "75%": [media.quantile(q=0.75)]
    #  }
    
    data=pd.DataFrame()
    if not quartil:
        data["Soma"]=media.sum(axis=1)
        data["Min"]=media.min(axis=1)
        data["Max"]=media.max(axis=1)
        data["Media"]=media.mean(axis=1)
        data["Mediana"]=media.median(axis=1)
        data["Coef_Var"]=media.std(axis=1,ddof=0)/media.median(axis=1)
        data["Vari_N"]=media.var(axis=1, ddof=0)
        data["Vari_N-1"]=media.var(axis=1)
        data["DesvPad_N"]=media.std(axis=1, ddof=0)
        data["DesvPad_N-1"]=media.std(axis=1)
        data["25%"] = media.quantile(axis=1, q=0.25)
        data["50%"] = media.quantile(axis=1, q=0.50)
        data["75%"] = media.quantile(axis=1, q=0.75)
        return data 
    else:
        data = pd.DataFrame(data=media, columns=["Mortalidade"])
        quartil=pd.DataFrame()
        q1 =  media.quantile(q=0.25)
        q2 = media.quantile(q=0.50)
        q3 = media.quantile(q=0.75)
        for x in data.index:
             if data.loc[x,"Mortalidade"] <= q1:
                    data.loc[x, "Quartil"] = 1
             elif data.loc[x,"Mortalidade"] > q1 and data.loc[x,"Mortalidade"] <=q2: 
                    data.loc[x, "Quartil"] = 2
             elif data.loc[x,"Mortalidade"] > q2 and data.loc[x,"Mortalidade"] <=q3:
                    data.loc[x, "Quartil"] = 3
             else:
                data.loc[x, "Quartil"] = 4
        data['Quartil'] = data['Quartil'].astype(int)
        data = data.set_index(["Quartil"])
        quartil["Soma"] = data.groupby('Quartil').sum()
        quartil["Min"]=data.groupby('Quartil').min()
        quartil["Max"]=data.groupby('Quartil').max()
        quartil["Media"]=data.groupby('Quartil').mean()
        quartil["Mediana"]=data.groupby('Quartil').median()
        quartil["Coef_Var"]=data.groupby('Quartil').std(ddof=0)/data.groupby('Quartil').median()
        quartil["Vari_N"]=data.groupby('Quartil').var(ddof=0)
        quartil["Vari_N-1"]=data.groupby('Quartil').var()
        quartil["DesvPad_N"]=data.groupby('Quartil').std(ddof=0)
        quartil["DesvPad_N-1"]=data.groupby('Quartil').std()
        quartil["25%"] = data.groupby('Quartil').quantile(q=0.25)
        quartil["50%"] = data.groupby('Quartil').quantile(q=0.50)
        quartil["75%"] = data.groupby('Quartil').quantile(q=0.75)
    return quartil
```

```python
def outilers(media):
    data = pd.DataFrame(data=media, columns=["Mortalidade"])
    data.boxplot()
    plt.title("Media por municipio")
    plt.ylabel("taxa")
    plt.show()
    
    q1 = media.quantile(q=0.25)
    q3 = media.quantile(q=0.75)
    lim_inf = q1 - 1.5 * (q3-q1)
    lim_sup = q3 + 1.5 * (q3-q1)
    print(f"Limite Inferior: {lim_inf:,.3f}")
    print(f"Limite Superior: {lim_sup:,.3f}")
    
    outilers = [x for x in data.index if data.loc[x,"Mortalidade"] < lim_inf or data.loc[x,"Mortalidade"] > lim_sup]
    print(data)
    print(f"outilers: {outilers}")
```

```python
def outilers_dataframe(media):
    data = pd.DataFrame(data=media, columns=["Mortalidade"])
    q1 = data.quantile(q=0.25)
    q3 = data.quantile(q=0.75)
    lim_inf = q1 - 1.5 * (q3-q1)
    lim_sup = q3 + 1.5 * (q3-q1)
    filtro = (data.Mortalidade < float(lim_inf)) | (data.Mortalidade > float(lim_sup))
    return data.loc[filtro]
```
