```python
#Importa as bibliotecas necessárias
import pandas as pd
import openpyxl

#Configura Pandas para mostrar todas as linhas
pd.set_option('display.max_rows', 20)
```

```python
#Carrega o dataframe de dados
df = pd.read_excel('tx_bruta_mortalidade_RJ.xlsx')
df=df.set_index("Município")
```

```python
#Exibe informações por Ano

df_Statisticas=pd.DataFrame()

df_Statisticas["Soma"]=df.sum()

df_Statisticas["Min"]=df.min()
df_Statisticas["Max"]=df.max()

df_Statisticas["Media"]=df.mean()
df_Statisticas["Mediana"]=df.median()
df_Statisticas["Kurtose"]=df.kurtosis()
df_Statisticas["Coef_Assi"]=df.skew()

df_Statisticas["Coef_Var"]=df.std(ddof=0)/df.median()

df_Statisticas["Vari_N"]=df.var(ddof=0)
df_Statisticas["Vari_N-1"]=df.var()
df_Statisticas["DesvPad_N"]=df.std(ddof=0)
df_Statisticas["DesvPad_N-1"]=df.std()

df_Statisticas["10%"]=df.quantile(.1)
df_Statisticas["20%"]=df.quantile(.2)
df_Statisticas["25%"]=df.quantile(.3)
df_Statisticas["50%"]=df.quantile(.4)
df_Statisticas["75%"]=df.quantile(.4)
df_Statisticas["90%"]=df.quantile(.4)

df_Statisticas
```

```python
#Exibe informações por Município

df_Municipios=pd.DataFrame()

df_Municipios["Soma"]=df.sum(axis=1)
df_Municipios["Media"]=df.mean(axis=1)

df_Municipios["Min"]=df.min(axis=1)
df_Municipios["Max"]=df.max(axis=1)

df_Municipios["Media"]=df.mean(axis=1)
df_Municipios["Mediana"]=df.median(axis=1)
df_Municipios["Kurtose"]=df.kurtosis(axis=1)
df_Municipios["Coef_Assi"]=df.skew(axis=1)

df_Municipios["Coef_Var"]=df.std(axis=1,ddof=0)/df.median(axis=1)

df_Municipios["Vari_N"]=df.var(axis=1,ddof=0)
df_Municipios["Vari_N-1"]=df.var(axis=1)
df_Municipios["DesvPad_N"]=df.std(axis=1,ddof=0)
df_Municipios["DesvPad_N-1"]=df.std(axis=1)

df_Municipios.sort_values(by="Media",ascending=False)
```

```python
#Estatisticas Normalizadas
normalizado_Statisticas=(df_Statisticas-df_Statisticas.min())/(df_Statisticas.max()-df_Statisticas.min())
normalizado_Statisticas
```

```python
#Municipios Normalizadas
normalizado_Municipios=(df_Municipios-df_Municipios.min())/(df_Municipios.max()-df_Municipios.min())
normalizado_Municipios
```
