### **DF_MUN_2021 - Medidas de posição -Dispersão data frame com as médias dos municipios** ###

```python
# ----------------------------------------
# Dataframe df_mun_2021 92 x 2
# ----------------------------------------
# Medidas de posição / Dispersão
# ----------------------------------------
soma=df.sum()
print(soma)
```

```python
soma  = df_mun_2021['2021'].sum
media = df_mun_2021['2021'].mean()
min   = df_mun_2021['2021'].min()
max   = df_mun_2021['2021'].max()
mediana   = df_mun_2021['2021'].median()
moda1 = df_mun_2021['2021'].mode(1) # Função moda no pandas
moda1.shape # Cada municipio tem uma moda (92)
n=len(moda1)
print('n=',n, '\nmoda1',moda1)
moda2 = stats.mode(df_mun_2021['2021']) # função moda no scipy

var = df_mun_2021['2021'].var() # Variância amaostral (padrão)
dp = df_mun_2021['2021'].std() # desvio padrão amostral
cv = dp/media  # cv = dp/média (amostral)
```
